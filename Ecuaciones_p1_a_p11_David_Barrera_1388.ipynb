Ecuación: $x + 8 = 15$Explicación: Este es un problema de suma simple. Para aislar la incógnita $x$, aplicamos la operación inversa a la suma, que es la resta. Restamos 8 en ambos lados de la igualdad, lo que nos da como resultado $x = 15 - 8$.

```python
from sympy import symbols, Eq, solve

x = symbols('x')
ecuacion = Eq(x + 8, 15)
resultado = solve(ecuacion, x)
print(f"La solución para x es: {resultado[0]}")
```

Problema 3: $3x + 4x = 21$Explicación: Al igual que en problemas anteriores, comenzamos agrupando los términos que contienen la misma variable. Sumamos los coeficientes de $x$: $3x + 4x$ nos da un total de $7x$. Ahora la ecuación es $7x = 21$. Para dejar a la $x$ sola, realizamos la operación contraria a la multiplicación del coeficiente 7, por lo que dividimos ambos lados entre 7, obteniendo así el valor de $x$.

```python
from sympy import symbols, Eq, solve
x = symbols('x')
ecuacion = Eq(3*x + 4*x, 21)
resultado = solve(ecuacion, x)
print(f"El valor de x es: {resultado[0]}")
```

Problema 5: $6a - 2a + a = 20$Explicación: Primero, simplificamos el lado izquierdo de la ecuación sumando y restando los coeficientes de $a$. Tenemos $6 - 2 + 1$, lo que da como resultado $5a$. Con la ecuación reducida a $5a = 20$, procedemos a despejar $a$ dividiendo el valor de la constante (20) entre el coeficiente de la variable (5). Este proceso nos permite encontrar el valor único que satisface la igualdad.

```python
from sympy import symbols, Eq, solve
a = symbols('a')
ecuacion = Eq(6*a - 2*a + a, 20)
resultado = solve(ecuacion, a)
print(f"El valor de a es: {resultado[0]}")
```

Problema 7: $4x + 3 = 2x + 11$Explicación: Para resolver esta ecuación, buscamos agrupar los términos variables y los términos constantes. Restamos $2x$ a ambos lados para concentrar las $x$ en el lado izquierdo ($4x - 2x = 2x$). Luego, restamos 3 en ambos lados para trasladar las constantes al lado derecho ($11 - 3 = 8$). Finalmente, dividimos 8 entre 2 para despejar la incógnita.

```python
from sympy import symbols, Eq, solve
x = symbols('x')
ecuacion = Eq(4*x + 3, 2*x + 11)
resultado = solve(ecuacion, x)
print(f"El valor de x es: {resultado[0]}")
```

Problema 9: $9m + 4 = 5m + 20$Explicación: Iniciamos el proceso restando $5m$ en ambos miembros de la ecuación, lo cual nos reduce el lado izquierdo a $4m + 4 = 20$. El siguiente paso lógico es eliminar la constante del lado izquierdo restando 4 en ambos lados, dejándonos con la igualdad $4m = 16$. Para finalizar, dividimos 16 entre 4, lo cual nos da el resultado preciso de la variable $m$.

```python
from sympy import symbols, Eq, solve
m = symbols('m')
ecuacion = Eq(9*m + 4, 5*m + 20)
resultado = solve(ecuacion, m)
print(f"El valor de m es: {resultado[0]}")
```

Problema 11: $6a + 5 = 2a + 21$Explicación: El procedimiento consiste en trasladar las incógnitas a la izquierda y los números a la derecha. Restamos $2a$ de ambos lados ($6a - 2a = 4a$) y restamos 5 de ambos lados ($21 - 5 = 16$). Esto simplifica la ecuación original a $4a = 16$. Para concluir el despeje, dividimos 16 entre 4, encontrando así el valor que equilibra ambos lados de la expresión original.

```python
from sympy import symbols, Eq, solve
a = symbols('a')
ecuacion = Eq(6*a + 5, 2*a + 21)
resultado = solve(ecuacion, a)
print(f"El valor de a es: {resultado[0]}")
```
