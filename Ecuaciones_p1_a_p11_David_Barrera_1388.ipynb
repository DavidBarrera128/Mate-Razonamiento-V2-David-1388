Problema 1: $x + 8 = 15$Explicación: Este es un problema de suma simple. Para aislar la incógnita $x$, aplicamos la operación inversa a la suma, que es la resta. Restamos 8 en ambos lados de la igualdad, lo que nos da como resultado $x = 15 - 8$.

```python
from sympy import symbols, Eq, solve

x = symbols('x')
# Para la ecuación x + b = c
b = int(input("Ingrese el valor para 'b' (el número que suma o resta a x, ej. 8): "))
c = int(input("Ingrese el valor para 'c' (el número al otro lado de la igualdad, ej. 15): "))
ecuacion = Eq(x + b, c)
resultado = solve(ecuacion, x)
print(f"La solución para x es: {resultado[0]}")
```

Problema 3: $3x + 4x = 21$Explicación: Al igual que en problemas anteriores, comenzamos agrupando los términos que contienen la misma variable. Sumamos los coeficientes de $x$: $3x + 4x$ nos da un total de $7x$. Ahora la ecuación es $7x = 21$. Para dejar a la $x$ sola, realizamos la operación contraria a la multiplicación del coeficiente 7, por lo que dividimos ambos lados entre 7, obteniendo así el valor de $x$.

```python
from sympy import symbols, Eq, solve
x = symbols('x')
# Para la ecuación coeff1*x + coeff2*x = constant
coeff1 = int(input("Ingrese el primer coeficiente de x (ej. 3): "))
coeff2 = int(input("Ingrese el segundo coeficiente de x (ej. 4): "))
constant = int(input("Ingrese el valor constante (ej. 21): "))
ecuacion = Eq(coeff1*x + coeff2*x, constant)
resultado = solve(ecuacion, x)
print(f"El valor de x es: {resultado[0]}")
```

Problema 5: $6a - 2a + a = 20$Explicación: Primero, simplificamos el lado izquierdo de la ecuación sumando y restando los coeficientes de $a$. Tenemos $6 - 2 + 1$, lo que da como resultado $5a$. Con la ecuación reducida a $5a = 20$, procedemos a despejar $a$ dividiendo el valor de la constante (20) entre el coeficiente de la variable (5). Este proceso nos permite encontrar el valor único que satisface la igualdad.

```python
from sympy import symbols, Eq, solve
a = symbols('a')
# Para la ecuación coeff1*a + coeff2*a + coeff3*a = constant
coeff1 = int(input("Ingrese el primer coeficiente de 'a' (ej. 6): "))
coeff2 = int(input("Ingrese el segundo coeficiente de 'a' (ej. -2): "))
coeff3 = int(input("Ingrese el tercer coeficiente de 'a' (ej. 1): "))
constant = int(input("Ingrese el valor constante (ej. 20): "))
ecuacion = Eq(coeff1*a + coeff2*a + coeff3*a, constant)
resultado = solve(ecuacion, a)
print(f"El valor de a es: {resultado[0]}")
```

Problema 7: $4x + 3 = 2x + 11$Explicación: Para resolver esta ecuación, buscamos agrupar los términos variables y los términos constantes. Restamos $2x$ a ambos lados para concentrar las $x$ en el lado izquierdo ($4x - 2x = 2x$). Luego, restamos 3 en ambos lados para trasladar las constantes al lado derecho ($11 - 3 = 8$). Finalmente, dividimos 8 entre 2 para despejar la incógnita.

```python
from sympy import symbols, Eq, solve
x = symbols('x')
# Para la ecuación coeff_x1*x + const_left = coeff_x2*x + const_right
coeff_x1 = int(input("Ingrese el coeficiente de 'x' en el lado izquierdo (ej. 4): "))
const_left = int(input("Ingrese la constante en el lado izquierdo (ej. 3): "))
coeff_x2 = int(input("Ingrese el coeficiente de 'x' en el lado derecho (ej. 2): "))
const_right = int(input("Ingrese la constante en el lado derecho (ej. 11): "))
ecuacion = Eq(coeff_x1*x + const_left, coeff_x2*x + const_right)
resultado = solve(ecuacion, x)
print(f"El valor de x es: {resultado[0]}")
```

Problema 9: $9m + 4 = 5m + 20$Explicación: Iniciamos el proceso restando $5m$ en ambos miembros de la ecuación, lo cual nos reduce el lado izquierdo a $4m + 4 = 20$. El siguiente paso lógico es eliminar la constante del lado izquierdo restando 4 en ambos lados, dejándonos con la igualdad $4m = 16$. Para finalizar, dividimos 16 entre 4, lo cual nos da el resultado preciso de la variable $m$.

```python
from sympy import symbols, Eq, solve
m = symbols('m')
# Para la ecuación coeff_m1*m + const_left = coeff_m2*m + const_right
coeff_m1 = int(input("Ingrese el coeficiente de 'm' en el lado izquierdo (ej. 9): "))
const_left = int(input("Ingrese la constante en el lado izquierdo (ej. 4): "))
coeff_m2 = int(input("Ingrese el coeficiente de 'm' en el lado derecho (ej. 5): "))
const_right = int(input("Ingrese la constante en el lado derecho (ej. 20): "))
ecuacion = Eq(coeff_m1*m + const_left, coeff_m2*m + const_right)
resultado = solve(ecuacion, m)
print(f"El valor de m es: {resultado[0]}")
```

Problema 11: $6a + 5 = 2a + 21$Explicación: El procedimiento consiste en trasladar las incógnitas a la izquierda y los números a la derecha. Restamos $2a$ de ambos lados ($6a - 2a = 4a$) y restamos 5 de ambos lados ($21 - 5 = 16$). Esto simplifica la ecuación original a $4a = 16$. Para concluir el despeje, dividimos 16 entre 4, encontrando así el valor que equilibra ambos lados de la expresión original.

```python
from sympy import symbols, Eq, solve
a = symbols('a')
# Para la ecuación coeff_a1*a + const_left = coeff_a2*a + const_right
coeff_a1 = int(input("Ingrese el coeficiente de 'a' en el lado izquierdo (ej. 6): "))
const_left = int(input("Ingrese la constante en el lado izquierdo (ej. 5): "))
coeff_a2 = int(input("Ingrese el coeficiente de 'a' en el lado derecho (ej. 2): "))
const_right = int(input("Ingrese la constante en el lado derecho (ej. 21): "))
ecuacion = Eq(coeff_a1*a + const_left, coeff_a2*a + const_right)
resultado = solve(ecuacion, a)
print(f"El valor de a es: {resultado[0]}")
```
