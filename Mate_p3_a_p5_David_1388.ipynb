Problema 3:

Resta básicaPlanteamiento: Un número desconocido ($x$) pierde 7 unidades y el resultado es 18.Ecuación:$$x - 7 = 18$$Despeje: Para encontrar $x$, movemos el $-7$ al otro lado con la operación contraria (suma).$$x = 18 + 7$$Solución:$$x = 25$$

```python
# Problema 3
# Definimos los valores conocidos
resultado = int(input("Introduce el resultado (por ejemplo, 18): "))
resta = int(input("Introduce la cantidad restada (por ejemplo, 7): "))

# Para hallar x en x - 7 = 18, sumamos: x = 18 + 7
numero_buscado = resultado + resta

print(f"Resultado Problema 3: {numero_buscado}")
```

Problema 4:

Ecuación lineal de dos pasosPlanteamiento: Tenemos el triple de un número ($3x$) y le añadimos 4 ($+ 4$), lo que nos da 31.Ecuación:$$3x + 4 = 31$$Despeje:Primero, quitamos lo que está sumando:$$3x = 31 - 4$$Resolvemos la resta:$$3x = 27$$Ahora, quitamos el 3 que multiplica a la $x$ pasándolo a dividir:$$x = \frac{27}{3}$$Solución:$$x = 9$$

```python
# problema 4
# Definimos los componentes de 3x + 4 = 31
total = int(input("Introduce el total (por ejemplo, 31): "))
suma = int(input("Introduce el valor a sumar (por ejemplo, 4): "))
multiplo = int(input("Introduce el valor por el que se multiplica (por ejemplo, 3): "))

# Primero restamos el valor constante y luego dividimos por el coeficiente
x = (total - suma) / multiplo

print(f"Resultado Problema 4: {int(x)}")
```

Problema 5:

 Perímetro de un terreno (Geometría)Planteamiento: * Un terreno rectangular tiene Ancho ($W$) y Largo ($L$).El problema dice que el largo mide 5 metros más que el ancho: $L = W + 5$.El perímetro ($P$) es la suma de los cuatro lados ($2L + 2W$) y es igual a 50.Ecuación inicial:$$2L + 2W = 50$$Sustitución: Reemplazamos $L$ por $(W + 5)$:$$2(W + 5) + 2W = 50$$Despeje:Multiplicamos el paréntesis:$$2W + 10 + 2W = 50$$Sumamos las $W$:$$4W + 10 = 50$$Restamos 10:$$4W = 40$$Dividimos entre 4:$$W = 10$$Solución:Ancho ($W$): 10 metros.Largo ($L$): $10 + 5 =$ 15 metros.

```python
# Prblema 5
# Perímetro total = 50. La fórmula es 2*(W + 5) + 2*W = 50 -> 4W + 10 = 50
perimetro_total = int(input("Introduce el perímetro total (por ejemplo, 50): "))
diferencia_largo = int(input("Introduce la diferencia de largo (por ejemplo, 5): "))

# Despejando W: W = (Perímetro - 2*Diferencia) / 4
ancho = (perimetro_total - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"Resultado Problema 5:")
print(f"El ancho mide: {ancho} metros")
print(f"El largo mide: {largo} metros")
```
